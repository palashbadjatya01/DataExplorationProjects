```python
import pandas as pd
import numpy as np
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
import bqplot
```

```python
#Reading buiding data set in building variable also replacing 0 values

buildings = pd.read_csv("https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv",
            na_values = {'Square Footage': 0,
                       'Year Acquired': 0,
                       'Year Constructed': 0}) 

buildings
```

```python
#Checking the shape of the buildings variable
buildings.shape
```

```python
#Just to double check the values on plots
criteria = buildings.groupby("County")[["Square Footage"]].describe()
criteria
```

```python
#Using pivot tables to prepare data for heat map and also replacing NaN with 0 

columns_example = pd.pivot_table(buildings, index = ['Agency Name'], columns = ['County'], values = ['Square Footage'], aggfunc = np.mean)
columns_example.fillna(0)
```

```python
# 1. Data -- same as above

# Tried using styles from matpltlib
# def make_plot(Style):
#     with plt.style.context(Style):

# 2. Scales
col_sc = bqplot.ColorScale(scheme = 'Blues') # bqplot will use blue as color scheme
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# 3. Marks
heat_map = bqplot.GridHeatMap(color=columns_example.values, scales={'color':col_sc, 'row':x_sc, 'column':y_sc},
                             interactions={'click':'select'},
                             anchor_style={'fill':'blue'})

# 4. Axis 
# c_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale=x_sc)
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical')

# 5. Interactions
def on_selected(change): # take a look at what changes when I select a new bin
    i,j = change['owner'].selected[0]
    mask = (buildings['Agency Name'] == columns_example.index[i]) & \
            (buildings['County'] == columns_example.columns.levels[1][j])
    # grab buildings subset dataframe
    buildings_subset = buildings[mask]
    # grab years and sqft from pandas
    grouped = buildings_subset.groupby("Year Acquired")["Square Footage"].sum()
    years = grouped.index
    sqft = grouped.values
    bar.x = years
    bar.y = sqft

heat_map.observe(on_selected,'selected')

# 6. Figure object
# pz = bqplot.interacts.PanZoom(scales={'x':[x_sc], 'y':[y_sc]})
fig = bqplot.Figure(marks= [heat_map], axes= [ x_ax, y_ax],)
display(fig)
   
#Styles code but did not worked well    
# #Radio button to select style of the plot
# Style= ipywidgets.Dropdown( options= ['seaborn-talk', 'ggplot', 'classic', 'tableau-colorblind10', 'Solarize_Light2'],
#                     value='seaborn-talk', #default value
#                     description='Style:', #name of radiobutton
#                     disabled=False
#                 )
# # created interaction of all above inputs with the function
# ipywidgets.interact(make_plot, Style = Style)
```

```python
# (1) scales: x/y, linear

x_sch = bqplot.LinearScale() # range of z-axis data
y_sch = bqplot.LinearScale() # frequency of z-axis data in bins

# (2) axis: x/y
x_axh = bqplot.Axis(scale = x_sch, label = 'Year')
y_axh = bqplot.Axis(scale = y_sch, 
                    orientation = 'vertical', 
                    label='Total Sq Footage')

# plot
bar = bqplot.Bars(x = years, y = sqft,
                    scales = {'x': x_sch, 'y': y_sch})

fig_bar = bqplot.Figure(marks = [bar], axes = [x_axh, y_axh])
display(fig_bar)
```
